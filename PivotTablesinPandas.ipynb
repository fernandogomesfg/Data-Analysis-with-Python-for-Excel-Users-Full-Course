# The pivot_table() method

```python
# read and show dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(index = "Gender", aggfunc = "sum") # retornando dados da tabela de vendas em categoria de genero
```

```python
# make a pivot table, add an aggregate function and select some columns
df_sales.pivot_table(index = "Gender", 
               values = ["Quantity", "Total" ],
               aggfunc = "sum")
```

```python
# make how much male and female spend in each produt
df_sales.pivot_table(index = "Gender", 
                     columns = "Product line",
                     values = ["Quantity", "Total" ],
                     aggfunc = "sum") # Gastos masculinos e femininos em cada tipo de produto
```

```python
 # Gastos masculinos e femininos em cada tipo de produto incluindo a cidade de cada categoria de genero
df_sales.pivot_table(index = "Gender", 
                     columns = ["Product line", "City"],
                     values = ["Quantity", "Total"],
                     aggfunc = "sum")
```
