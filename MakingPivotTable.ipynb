```python
import pandas as pd
```

```python
df_population_total = pd.read_csv('population_total.csv')
```

```python
# dropping null values
df_population_total = df_population_total.dropna()
```

```python
# making a pivot table
df_pivot = df_population_total.pivot(index = 'year', columns = 'country', values = 'population')
```

```python
df_pivot
```

```python
# select some contries
df_pivot = df_pivot[['Mozambique', 'Canada', 'China', 'India', 'Brazil']]
```

```python
df_pivot
```

 # Lineplot - permite plotar grafios de linha

```python
df_pivot.plot(kind = 'line')  #  evolucao da populacao durante esses anos
```

```python
df_pivot.plot(kind = 'line', xlabel = 'Ano', ylabel = 'Populacao', title = 'Populacao (1955 - 2020)' )  #  evolucao da populacao durante esses anos com alguns argumento acrescentados
```

```python
df_pivot.plot(kind = 'line', xlabel = 'Ano', ylabel = 'Populacao', title = 'Populacao (1955 - 2020)', figsize = (12,6) ) 
```

# Barplot

```python
# selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# show  new dataframe
df_pivot_2020
```

```python
# traspose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# makintg barplot
df_pivot_2020.plot(kind = 'bar', color = 'orange', xlabel = 'Ano', ylabel = 'Populacao', title = 'Populacao (2020)', figsize = (12,6))
```
