```python
import pandas as pd
```

```python
df_population_total = pd.read_csv('population_total.csv')
```

```python
# dropping null values
df_population_total = df_population_total.dropna()
```

```python
# making a pivot table
df_pivot = df_population_total.pivot(index = 'year', columns = 'country', values = 'population')
```

```python
df_pivot
```

```python
# select some contries
df_pivot = df_pivot[['Mozambique', 'Canada', 'China', 'India', 'Brazil']]
```
