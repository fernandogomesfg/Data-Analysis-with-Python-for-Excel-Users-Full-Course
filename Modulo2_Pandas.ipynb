# Creating a dataframe from any array

```python
import pandas as pd
import numpy as np
```

## Option 1

```python
# creating an array
data = np.array([[1,4], [2,5], [3,6]])
```

```python
# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
```

```python
# showing the data frame
df
```

## Option 2

```python
# creating an array with list shape
data = [[1,4], [2,5], [3,6]]
```

```python
# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
```

```python
# showing the data frame
df
```

# Creating a DataFrame from a dictionary

```python
# lists used for this example
provinces = ['Maputo', 'Gaza', 'Inhambane', 'Manica']
population = [2764169, 1396024 , 1496824, 1911237]
```

```python
# Storing lists within a dictionary
dict_provinces = {'Provinces': provinces, 'Population': population}
```

```python
# creating the dataframe
df_population = pd.DataFrame(dict_provinces)
```

```python
# showing the data frame
df_population
```
